```python
# Importing our libraries/modules:
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import numpy as np
import os
import requests
from scipy.stats import linregress
import scipy.stats as st
```

```python
census_2017 = pd.read_csv('../Output/2017_cleaned_state.csv')
#census_2017.head()
census_2015 = pd.read_csv('../Output/2015_cleaned_state.csv')
# census_2015.head()
```

```python
#scatter compare unemployment with total population for each state
# x-axis Unemployment per state
# y-axis is total population
# unemployment = census_2017["Unemployment"]
# total_population = census_2017["TotalPop"]

x_values = census_2017["Unemployment"]
y_values = census_2017["TotalPop"]

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(x_values,regress_values,"r-")
print(f"{rvalue**2}")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")

plt.xlabel("Unemployment")
plt.ylabel("Total Population (by million)")
plt.title("Unemployment vs Total Population per State")
plt.scatter(x_values, y_values, marker = "D")
plt.show()
#print(total_population)
plt.scatter
```

```python
#Observation -The appearance of a correlation exists that suggests Total population and unemployment are impacted. The higher the population the more risk of unemployment. 
```
