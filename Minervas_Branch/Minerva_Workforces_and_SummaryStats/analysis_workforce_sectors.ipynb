```python
# Importing our libraries/modules:
import pandas as pd
import scipy
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading our cleaned dataset '2015' and printing for preview:
states_2015 = pd.read_csv('../Output/2015_cleaned_state.csv')
states_2015.head()
```

```python
# Building a dataframe to find the types that makes up the states workforce based on 2015:
workforce_2015 = states_2015[['State', 'TotalPop', 'Professional', 'Service', 'Office', 'Construction', 'Production']]
workforce_2015
```

```python
# Creating a 'Stacked Bar Graph' for 2015 to display the ratios of workforce sectors:
x = workforce_2015['State']
y0 = workforce_2015['Professional']
y1 = workforce_2015['Service']
y2 = workforce_2015['Office']
y3 = workforce_2015['Construction']
y4 = workforce_2015['Production']

plt.figure(figsize=(15, 15))

p1 = plt.bar(x,y0, color='blue', label='Professional')
p2 = plt.bar(x,y2, color='yellow', label='Office')
p3 = plt.bar(x,y1, color='red', label='Service')
p4 = plt.bar(x,y4, color='green', label='Production')
p5 = plt.bar(x,y3, color='orange', label='Construction')

plt.xlim(-0.75, len(x)-0.25)
plt.ylim(0, max(y0)+100000)

plt.title("Workforce Sectors based on 2015 Total State Population")
plt.xlabel("U.S. States, DC, & Puerto Rico")
plt.ylabel("Total Population per State")
plt.legend(loc='best')

tick_locations = [value for value in x]
plt.xticks(tick_locations, rotation=90)

plt.savefig('Images/Fig1.png')
plt.show()
```

#### This data visulation on 'Workforce Types' is based on 2015 Total State Population. This tells us that the workforce type is the professional industry is the largest workforce across most states. We can also determine that the service industry is also dependent on the population. More people more services needed. Another interesting find is that you can tell the US outsources a lot of the production work because there is hardly any presence in most states to the exception of the few like Illinois, Indiana, Michigan, and Ohio.

```python
# Building a dataframe to find the workforce types that make-up the states based on 2015:
types_2015 = states_2015[['State', 'TotalPop', 'PrivateWork', 'PublicWork', 'SelfEmployed', 'FamilyWork']]
types_2015
```

```python
# Converting our numbers to percentages for easier plotting:
total_p_15 = types_2015['TotalPop']
pct_private_15 = (types_2015['PrivateWork'] / total_p_15) * 100
pct_public_15 = (types_2015['PublicWork'] / total_p_15) * 100
pct_self_15 = (types_2015['SelfEmployed'] / total_p_15) * 100
pct_family_15 = (types_2015['FamilyWork'] / total_p_15) * 100
```

```python
# Creating a 'Stacked Bar Graph' for 2015 to display the ratios of workforce sectors:
x = types_2015['State']

y0 = pct_private_15
y1 = pct_public_15
y2 = pct_family_15
y3 = pct_self_15

plt.figure(figsize=(15, 15))

p1 = plt.bar(x,y0, color='blue', label='PrivateWork')
p2 = plt.bar(x,y1, color='green', label='PublicWork')
p3 = plt.bar(x,y2, color='red', label='FamilyWork')
p4 = plt.bar(x,y3, color='orange', label='SelfEmployed')


plt.xlim(-0.75, len(x)-0.25)
plt.ylim(0, max(y0)+1)

plt.title("Workforce (types) based on 2015 Total State Population")
plt.xlabel("U.S. States, DC, & Puerto Rico")
plt.ylabel("Total Population per State (%)")
plt.legend(loc='best')

tick_locations = [value for value in x]
plt.xticks(tick_locations, rotation=90)

plt.savefig('Images/Fig3.png')
plt.show()
```
