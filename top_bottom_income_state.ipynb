## 2015 Census Data

```python
# Reading our dataset 'census_2015.csv' and printing a preview:
al_census_2015 = pd.read_csv('Output/2015_cleaned_state.csv')
#al_census_2015.head()
```

```python
al_2015_narrowed = al_census_2015[['State','TotalPop','IncomePerCap',
                                  'Employed','Drive','Carpool',
                                  'Transit','Walk','OtherTrans','MeanCommute','Unemployment']].copy()
```

```python
#sort data set by highest income per capita to find the top 5 
al_2015_narrowed.sort_values('IncomePerCap',ascending=False,inplace=True)
#al_2015_narrowed.head(5)
```

```python
#Create a new df with only the states with the top 5 income per capita for 2015
al_top_15 = al_2015_narrowed.loc[(al_2015_narrowed['IncomePerCap']>=35554)].copy()
al_top_15.sort_values('IncomePerCap',ascending=False,inplace=True)
al_top_15
```

```python
#sort data set by lowest income per capita to find the bottom 5 
al_2015_narrowed.sort_values('IncomePerCap',ascending=True,inplace=True)
al_2015_narrowed.head(5)
```

```python
#Create a new df with only the states with the bottom 5 income per capita
al_bottom_15 = al_2015_narrowed.loc[(al_2015_narrowed['IncomePerCap']<=20700)].copy()
al_bottom_15.sort_values('IncomePerCap',ascending=True,inplace=True)
al_bottom_15
```

```python
#sort data set by highest income per capita to find the top 5 
al_2017_narrowed.sort_values('IncomePerCap',ascending=False,inplace=True)
al_2017_narrowed.head(5)
```
